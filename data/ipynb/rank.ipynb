```python
import pandas as pd
```

```python
df = pd.read_csv("./merged_file.csv")
df
```

```python
df['libraryCountRank'] = df['libraryCount'].rank(method='min', ascending=False) # 같은 값이 있을 경우 같은 순위를 부여하고, 다음 값은 순위가 중복된 자료 개수 다음 순위를 부여 예. (1, 2, 2, 2, 5)
df['busStationRank'] = df['busStation'].rank(method='min', ascending=False)
df['cultureCountRank'] = df['cultureCount'].rank(method='min', ascending=False)
df['medicalCountRank'] = df['medicalCount'].rank(method='min', ascending=False)
df['crimeRateRank'] = df['crimeRate'].rank(method='min', ascending=False)
df['youthRateRank'] = df['youthRate'].rank(method='min', ascending=False)
df['supermarketRank'] = df['supermarket'].rank(method='min', ascending=False)
df['parkRateRank'] = df['parkRate'].rank(method='min', ascending=False)
df['academyCountRank'] = df['academyCount'].rank(method='min', ascending=False)
df
```

```python
df.to_csv('./merged_file_rank.csv', index=False, encoding='utf-8-sig')
```

```python
df1 = df.groupby('gu').sum().drop(['dong', 'index', 'id', 'jeonseDeposit', 'monthDeposit', 'monthRent','academyCount','crimeRate','parkRate'], axis=1).rank(method='min', ascending=False)
df1
```

```python
df2 = pd.read_csv('./gu_info.csv')
df2.set_index('Region', inplace=True) # index 설정을 해줘야 제대로 값을 넣어줄 수 있음
df2['libraryCoutRank'] = df1['libraryCount']
df2['busStationRank'] = df1['busStation']
df2['cultureCountRank'] = df1['cultureCount']
df2['medicalCountRank'] = df1['medicalCount']
df2['youthRateRank'] = df1['youthRate']
df2['supermarketRank'] = df1['supermarket']
df2
```

```python
df2.to_csv('./gu_rank.csv', encoding='utf-8-sig')
```
