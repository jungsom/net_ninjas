## 사설학원 통계

http://data.seoul.go.kr/dataList/195/S/2/datasetView.do

각 자치구의 사설학원 중 평생직업 교육학원 소계를 dictionary 형태 `{'강남구': '493', '서초구': '278', '마포구': '188', '종로구': '142', ...}`로 정리한다.

csv 파일 읽어오기

```python
seoul_academies = pd.read_csv('/content/seoul_academy.csv')
print(seoul_academies)
print(seoul_academies.columns)  # 열 이름 추출
```

자치구명 추출하고 딕셔너리화

```python
seoul_regions = seoul_academies['자치구별(2)'][5:]  # 원본에서 자치구명이 있는 '자치구별(2)'열의 5~마지막행 추출
print(seoul_regions)

seoul_regions_dic = {}
for region in seoul_regions:
  seoul_regions_dic[region] = 0
print(seoul_regions_dic)  # {'종로구': 0, '중구': 0, '용산구':0, ...}
```

특정 자치구의 평생직업 교육학원 소계 (예: 종로구)



```python
# 5번째 행 '2022.9'열 셀 값 추출
seoul_jongno_academies = seoul_academies.loc[5, '2022.9']
print(seoul_jongno_academies)
```

반복문을 이용한 각 자치구의 평생직업 교육학원 소계

```python
# 각 자치구별 평생직업 교육학원 소계
for i in range(5, 30):
  seoul_region_academies = seoul_academies.loc[i, '2022.9']
  seoul_regions_dic[seoul_academies.loc[i, '자치구별(2)']] = seoul_region_academies

sorted_seoul_regions_dic = dict(sorted(seoul_regions_dic.items(), key=lambda item: int(item[1]), reverse=True))  # 학원 많은 자치구 순으로 정렬
print(sorted_seoul_regions_dic)
```

# 공공도서관 통계

https://lib.seoul.go.kr/slibsrch/main

1.   각 자치구의 공공도서관 개수를 dictionary 형태 `{'강남구': 16, '성북구': 13, '구로구': 12, ...}`로 정리한다.
2.   각 자치구 각 동의 공공도서관 개수를 dictionary 형태 `{'강남구': {'일원동': 3, '대치동': 3, '자곡동': 2, '역삼동': 2, '청담동': 2, '논현동': 1, '도곡동': 1, '삼성동': 1, '개포동': 1}, '성북구': {'돈암동': 3, '길음동': 2, '하월곡동': 2, '정릉동': 2, '상월곡동': 1, '장위동': 1, '종암동': 1}, ...}`로 정리한다.

```python
seoul_libraries = pd.read_excel('/content/seoul_library_B.xlsx', engine='openpyxl')
print(seoul_libraries)
print(seoul_libraries.columns)  # 열 이름 추출
```

```python
seoul_regions = seoul_libraries['Unnamed: 7'][3:]  # 원본에서 자치구명이 있는 'Unnamed: 7'열의 3~마지막행 추출
print(seoul_regions)

seoul_regions_dic = {}
for region in seoul_regions:
  seoul_regions_dic[region] = 0
print(seoul_regions_dic)  # {'종로구': 0, '중구': 0, '용산구':0, ...}
```

각 자치구의 공공도서관 개수

*   도서관 주소 목록에서 특정 자치구명이 반복해서 등장하는 횟수를 계산한다.

```python
gus = seoul_libraries['Unnamed: 7'][3:]  # 원본에서 자치구명이 있는 'Unnamed: 7'열의 5~마지막행 추출
gus_libraries = gus.value_counts() # 'Unnamed: 7'열에서 각 자치구명이 몇 번 등장하는지 계산
gus_libraries_dict = gus_libraries.to_dict()  # 딕셔너리화
print(gus_libraries_dict)
```

특정 자치구 동의 공공도서관 개수

```python
gu_dong_addresses = seoul_libraries['Unnamed: 11'][3:]  # 원본에서 공공도서관 주소가 있는 'Unnamed: 11'열의 3~마지막행 추출
gu_dong_libraries_dict = gus_libraries_dict # 각 자치구의 공공도서관 개수 딕셔너리 복사


import re
from collections import Counter


for key in gus_libraries_dict.keys():
  specific_gu_dong_addresses = gu_dong_addresses[gu_dong_addresses.str.contains(key, na=False)] # 공공도서관 주소 목록에서 특정 자치구 주소 추출
  specific_gu_dong_addresses_list = specific_gu_dong_addresses.tolist()  # 리스트화

  specific_dong_list = [gu_dong_address.split()[2] for gu_dong_address in specific_gu_dong_addresses_list]   # 특정 자치구 주소에서 동 이름만 추출해 새로운 리스트에 저장
  specific_dong_frequency_dict = dict(Counter(specific_dong_list))  # 각 동의 빈도수(도서관 수)를 딕셔너리화
  specific_dong_frequency_dict = dict(sorted(specific_dong_frequency_dict.items(), key=lambda item: int(item[1]), reverse=True))  # 도서관 많은 동 순으로 정렬

  for gu in gu_dong_libraries_dict:
    gu_dong_libraries_dict[key] = specific_dong_frequency_dict  # {'자치구': {'동': 0, '동': 0, ...}, ...} 형태로 딕셔너리화


print(gu_dong_libraries_dict)
```

```python
# 주어진 딕셔너리
data = gu_dong_libraries_dict

# 데이터를 변환하여 리스트화
rows = []
for key1, inner_dict in data.items():
    for key2, value in inner_dict.items():
        rows.append([key1, key2, value])

# 리스트를 데이터프레임으로 변환
df = pd.DataFrame(rows, columns=['구', '동', '빈도수'])

# 데이터프레임을 CSV 파일로 저장
df.to_csv('seoul_libraries_analysis.csv', index=False, encoding='utf-8-sig')

# 결과 출력
print(df)
```

# 버스 정류장 통계

```python
seoul_busstops = pd.read_csv('/content/seoul_busstop.csv')
print(seoul_busstops)
print(seoul_busstops.columns)  # 열 이름 추출
```

```python
seoul_busstops['자치구+동'] = seoul_busstops['자치구'] + ' ' + seoul_busstops['동']
print(seoul_busstops)
print(seoul_busstops.columns)  # 열 이름 추출
```

```python
gus = seoul_busstops['자치구'][0:]  # 원본에서 자치구명이 있는 'Unnamed: 7'열의 5~마지막행 추출
gu_busstops = gus.value_counts() # 'Unnamed: 7'열에서 각 자치구명이 몇 번 등장하는지 계산
gu_busstops_dict = gu_busstops.to_dict()  # 딕셔너리화
print(gu_busstops_dict)
```

```python
gu_dongs = seoul_busstops['자치구+동'][0:]  # 원본에서 자치구명이 있는 '자치구+동'열의 0~마지막행 추출

gu_dong_busstops_dict = gu_busstops_dict # 각 자치구의 정류장 개수 딕셔너리 복사


import re
from collections import Counter


for key in gu_busstops_dict.keys():
  specific_gu_dongs = gu_dongs[gu_dongs.str.contains(key, na=False)] # 자치구+동 목록에서 특정 자치구 이름이 포함된 값 추출
  specific_gu_dong_list = specific_gu_dongs.tolist()  # 리스트화

  # 삭제하고자 하는 특정 단어들
  remove_words = [key]

  # 리스트의 각 값에서 특정 단어와 숫자를 삭제하는 함수
  def remove_words_and_numbers(text, words):
      # 특정 단어들 삭제
      for word in words:
          text = re.sub(r'\b' + re.escape(word) + r'\b', '', text)
      # 양쪽 공백 및 다중 공백 제거
      text = re.sub(r'\s+', ' ', text).strip()
      return text

  for gu_dong in gu_dongs:
    if key in gu_dong:
      specific_dong_list = [remove_words_and_numbers(gu_dong, remove_words) for gu_dong in specific_gu_dong_list]   # 특정 자치구+동 값에서 자치구 이름 제거하고 새로운 리스트에 저장
      specific_dong_frequency_dict = dict(Counter(specific_dong_list))  # 각 동의 빈도수(정류장 수)를 딕셔너리화
      specific_dong_frequency_dict = dict(sorted(specific_dong_frequency_dict.items(), key=lambda item: int(item[1]), reverse=True))  # 정류장 많은 동 순으로 정렬

  for gu in gu_dong_busstops_dict:
    gu_dong_busstops_dict[key] = specific_dong_frequency_dict  # {'자치구': {'동': 0, '동': 0, ...}, ...} 형태로 딕셔너리화


print(gu_dong_busstops_dict)
```

```python
# 주어진 딕셔너리
data = gu_dong_busstops_dict

# 데이터를 변환하여 리스트화
rows = []
for key1, inner_dict in data.items():
    for key2, value in inner_dict.items():
        rows.append([key1, key2, value])

# 리스트를 데이터프레임으로 변환
df = pd.DataFrame(rows, columns=['구', '동', '빈도수'])

# 데이터프레임을 CSV 파일로 저장
df.to_csv('seoul_busstops_analysis.csv', index=False, encoding='utf-8-sig')

# 결과 출력
print(df)
```

# 대형마트 통계

```python
seoul_supermarkets = pd.read_csv('/content/seoul_supermarket.csv')
print(seoul_supermarkets)
print(seoul_supermarkets.columns)  # 열 이름 추출
```

```python
# 모든 셀 값의 두 번째 단어와 마지막 단어만 추출
seoul_supermarkets['second_and_last'] = seoul_supermarkets['도로명주소'].apply(lambda x: ' '.join([x.split()[1], x.split()[-1]]) if len(x.split()) >= 2 else None)
seoul_supermarkets['cleaned'] = seoul_supermarkets['second_and_last'].str.replace(r'[\(\)]', '', regex=True)
test = seoul_supermarkets['cleaned'].tolist()
print(seoul_supermarkets['cleaned'])
print(test)
```

```python
# 각 동의 빈도수(대형마트 수) 계산
seoul_supermarkets_counts = seoul_supermarkets['cleaned'].value_counts()

# value_counts() 결과를 데이터프레임으로 변환
seoul_supermarkets_counts_df = seoul_supermarkets_counts.reset_index()

# 열 이름 변경
seoul_supermarkets_counts_df.columns = ['구_동', '대형마트 수']

# 데이터프레임 확인
print(seoul_supermarkets_counts_df)
```

```python
# 두 단어로 분리하여 각각 새 열로 저장
seoul_supermarkets_counts_df[['구', '동']] = seoul_supermarkets_counts_df['구_동'].str.split(n=1, expand=True)

# 결과 확인
print(seoul_supermarkets_counts_df)

seoul_supermarkets_counts_df.to_csv('seoul_supermarket_analysis.csv', index=False, encoding='utf-8-sig')
```
